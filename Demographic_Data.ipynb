```python
import pandas as pd
import numpy as np
```

```python
demographicsData = pd.read_csv('Demographic_Data/DemoData.csv')
demographicsData.head()
```

```python
# Functions box

# create function to delete columns based on name
def deleteColumnNames(table, str):
    index = []
    for name in table.columns:
        if str in name:
            index.append(name)
    table = table.drop(columns = index, axis = 0)
    return table;
# created a function to create a table using specific columns from meta table
def grabSpecificColumns(table, str):
    index = ['Geographic Area Name']
    for name in table.columns:
        if str in name:
            index.append(name)
    table = table[index]
    return table

# convert dtype of columns to a specfic one other thatn Geographic Area name
def convertDtypes(table, str):
    #creating a list of columns to iterate through so I can turn strings into ints
    listOfColumns = table.columns.to_list()
    listOfColumns.remove('Geographic Area Name')
    table[listOfColumns] = table[listOfColumns].astype(str)
    return table
```

```python
# remove id column
demographicsData = deleteColumnNames(demographicsData, 'GEO_ID')

demographicsData.head()
```

```python
titleArray = []
for title in demographicsData.loc[0]:
    title = title.replace('!!', ' ')
    titleArray.append(title)
demographicsData.loc[0] = titleArray
demographicsData.head()
```

```python
demographicsData.columns = demographicsData.loc[0]
demographicsData = deleteColumnNames(demographicsData, 'Margin of Error')
demographicsData = demographicsData.drop(labels=0, axis=0)
```

```python
demographicsData.head()
```

```python
# The value (X) was the equivalent of NaN values in these tables so I removed them
index = []
for column in demographicsData:
    if '(X)' in demographicsData[column].loc[1]:
        index.append(column)
demographicsData = demographicsData.drop(columns=index, axis = 0)
demographicsData.head()
```

```python
# columns with N value had race information which is not what i needed in my analysis so I am removing those columns
index = []
for column in demographicsData:
    if 'N' in demographicsData[column].loc[1]:
        index.append(column)
demographicsData = demographicsData.drop(columns=index, axis = 0)
demographicsData.head()
```
