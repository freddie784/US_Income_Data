```python
import pandas as pd
```

```python
income_table = pd.read_csv('productDownload_2022-01-24T145502/State_by_State_Income_Info.csv')
```

```python
# create function to delete columns based on name
def deleteColumnNames(table, str):
    index = []
    for name in table.columns:
        if str in name:
            index.append(name)
    table = table.drop(columns = index, axis = 0)
    return table;
```

```python
income_table.head()
```

```python
# remove id column
income_table = deleteColumnNames(income_table, 'GEO_ID')

income_table.head()
```

```python
# replace !! with a space in the column names
rowOne =income_table.loc[0]
titleArray = []
for title in rowOne:
    title = title.replace('!!', ' ')
    titleArray.append(title)
income_table.loc[0] = titleArray
income_table.head()
```

```python
# replace column names with first row names
income_table.columns = income_table.loc[0]
```

```python
# drop duplicate column names
income_table = income_table.drop(labels=0, axis=0)
```

```python
income_table.head()
```

```python
# delete margin of error columns
noMarginofError = deleteColumnNames(income_table, 'Margin of Error')
noMarginofError.head()
```
